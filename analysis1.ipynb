# Sentiment Analysis of Tweets
Here we have segregated Covid related tweets and non-covid related tweets to find sentiment of the tweets.

```python
tweet_list1=[]
for tweet in tweets: 
    tweet_list1.append(tweet['full_text'])
    
tweet_list1
```

```python
import string
#removing punctuation from list using punctuation in string package and again storing changes in list
tweet_list1= [''.join(c for c in s if c not in string.punctuation) for s in tweet_list1]
```

```python
import re
#Removing numbers and special symbols using regular expressions and again storing changes in list
tweet_list1 = [re.sub('[^a-zA-Z]+', ' ', _) for _ in tweet_list1]
#converting all the data in merged_list to lower case 
tweet_list1=[w.lower() for w in tweet_list1]
```

```python
tweet_list1
```

```python
from nltk.tokenize import word_tokenize
#opeining stopwords file and reading all the data by splitting lines
with open(r'C:\Users\dnyap\Downloads\stopwords_en (2).txt') as f:
    stopwords = f.read().splitlines()
```

```python
#removing stopwords from list 
leng = len(tweet_list1)
#creating empty list to store the list elements without stopwords
list_without_sw=[]
#for loop to iterate through all the list
for i in range(0,leng):
    #taking i'th string text variable
    text=tweet_list1[i]
    #creating tokens of words from text string using word_tokenize in nltk
    text_tokens = word_tokenize(text)
    #iterating through every word in text_tokens and comparing it with words in stopwords list
    #if word does not matches with word in stopwords list then store it in tokens_without_sw
    tokens_without_sw = [word for word in text_tokens if not word in stopwords]
    #connecting all single words to create a complete string/sentence
    tt=" ".join(tokens_without_sw)
    #appending the strings without stopwords to list_without_sw
    list_without_sw.append(tt)
    
tweet_list1=list_without_sw
```

```python
#to find the tweets related to COVID-19 
covid_words=['covid', 'coronavirus', 'vaccine', 'vaccination', 'antibody', 'moderna', 'pfizer', 'johnson','wuhan','virus','china']
#empty list to store covid tweets
covid=[]
#creating empty list to store non covid tweets
non_covid=[]
#for loop to itearte through length of list
for i in range(0,len(tweet_list1)):
    #taking string in text
    text=tweet_list1[i]
    #if covid word found in tweets append to covid list
    if 'covid' in text:
        covid.append(text)
    #if coronavirus word found in tweets append to covid list
    elif 'coronavirus' in text:
        covid.append(text)
    #if vaccine word found in tweets append to covid list
    elif 'vaccine' in text:
        covid.append(text)
    #if vaccination word found in tweets append to covid list
    elif 'vaccination' in text:
        covid.append(text)
    #if moderna word found in tweets append to covid list
    elif 'moderna'  in text:
        covid.append(text)
    #if antibody  word found in tweets append to covid list
    elif 'antibody' in text:
        covid.append(text)
    #if pfizer word found in tweets append to covid list
    elif 'pfizer' in text:
        covid.append(text)
    #if cjohnson word found in tweets append to covid list
    elif 'johnson' in text:
        covid.append(text)
    elif 'wuhan' in text:
        covid.append(text)
    elif 'virus' in text:
        covid.append(text)
    elif 'china' in text:
        covid.append(text)
    #if no word found in tweets then append to non_covid list
    else:
        non_covid.append(text)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import defaultdict,OrderedDict
```

# Sentiment analysis of tweets not related to COVID

```python
#sentiment of non covid tweets
sentiment_obj = SentimentIntensityAnalyzer()

#finding snetiment score of non covid realted tweets and printing it using vader sentiment analyzer
score_non = sentiment_obj.polarity_scores(non_covid)
print("Sentiment for non COVID related tweets",score_non)

if score_non['compound'] >= 0.05 :
    print("Positive")
 
elif score_non['compound'] <= - 0.05 :
    print("Negative")
 
else :
    print("Neutral")
```
